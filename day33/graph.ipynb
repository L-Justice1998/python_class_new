```python
#set fig
import matplotlib.pyplot as plt
import random

x=range(2,28,2)
y=[random.randint(1,30) for i in x]
plt.figure(figsize=(20,10),dpi=80)
plt.plot(x,y)
plt.savefig('./g1.svg')
plt.show()
```

```python
#set ticks
x=range(1,24,2)
y=[random.randint(15,30) for i in x]
x_ticks_labels=["{}:00".format(i) for i in x]
y_ticks_labels=["{}".format(i) for
                i in range(min(y)-1,max(y)+1)]
plt.xticks(x,x_ticks_labels,rotation=45)
plt.yticks(range(min(y)-1,max(y)+1),y_ticks_labels)
#labels match tick
plt.plot(x,y)
plt.show()
```

```python
#set font
from matplotlib import font_manager

my_font=font_manager.FontProperties(fname='C:\Windows\Fonts\msyh.ttc'
                                    ,size=18)
x = range(0,120)
y = [random.randint(10,30) for i in range(120)]
plt.xlabel("",rotation=45,fontproperties=my_font)
plt.ylabel("",fontproperties=my_font)
plt.title("",fontproperties=my_font)
plt.plot(x,y)
plt.show()
```
