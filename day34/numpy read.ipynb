```python
import matplotlib.pyplot as plt
import numpy as np
us_file_path='./US_video_data_numbers.csv'
us_data=np.loadtxt(us_file_path,delimiter=',',dtype='int')

us_data_comment=us_data[:,-1]
us_data_comment=us_data_comment[us_data_comment<5000]
d=50
group_num=(max(us_data_comment)-min(us_data_comment))//d
plt.figure(figsize=(20,8),dpi=80)
plt.hist(us_data_comment,bins=group_num)
plt.xlabel("comment number")
plt.ylabel("video number")
plt.show()
```

```python
uk_data_path='./GB_video_data_numbers.csv'
uk_data=np.loadtxt(uk_data_path,delimiter=',')
uk_data=uk_data[uk_data[:,1]<500000]
uk_com=uk_data[:,1]
uk_like=uk_data[:,-1]
plt.scatter(uk_com,uk_like)
plt.xlabel("comment")
plt.ylabel("like")
plt.title("relation between comment and like")
plt.show()
```
